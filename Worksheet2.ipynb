```python
# Import required libraries
import pandas as pd

# Correct the file_path to match the Google Drive location used previously
file_path = "/content/drive/MyDrive/Datasets/medical_students_dataset.csv"

# Load the medical students dataset
medical_students_df = pd.read_csv(file_path)







```

```python
# Step 1: Inspect the dataset
print("Initial Dataset Info:")
print(medical_students_df.info())
```

```python
# Step 2: Identify and handle missing values

# Numerical columns - Fill with mean for continuous data
numerical_columns = ['Age', 'Height', 'BMI', 'Temperature', 'Heart Rate', 'Blood Pressure']
for col in numerical_columns:
    medical_students_df[col].fillna(medical_students_df[col].mean(), inplace=True)

# Interpolation for sequential IDs
medical_students_df['Student ID'] = medical_students_df['Student ID'].interpolate()

# Categorical column - Fill with mode
medical_students_df['Smoking'].fillna(medical_students_df['Smoking'].mode()[0], inplace=True)
```

```python
# Step 3: Handle remaining categorical missing values
# Fill missing values in categorical columns with 'Unknown' (or any domain-specific approach)
categorical_columns = ['Gender', 'Blood Type', 'Diabetes']
for col in categorical_columns:
    medical_students_df[col].fillna('Unknown', inplace=True)
```

```python
# Step 4: Check and manage duplicates
duplicates_count = medical_students_df.duplicated().sum()
medical_students_df.drop_duplicates(inplace=True)

```

```python
# Step 5: Summary after cleaning
print("\nCleaned Dataset Info:")
print(medical_students_df.info())
print(f"\nDuplicates Removed: {duplicates_count}")

```

```python
# Step 6: Save the cleaned dataset
cleaned_file_path = 'medical_students_cleaned.csv'  # Path for saving the cleaned dataset
medical_students_df.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved to: {cleaned_file_path}")
```

```python
# Problem 1: Subset and statistics
# Change 'titanic_diff' to 'df' to use the loaded DataFrame
subset_df = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
first_class_df = subset_df[subset_df['Pclass'] == 1]

# Statistics for 'Fare'
fare_stats = {
    "Mean": first_class_df['Fare'].mean(),
    "Median": first_class_df['Fare'].median(),
    "Max": first_class_df['Fare'].max(),
    "Min": first_class_df['Fare'].min(),
}

print("Fare Statistics:\n", fare_stats)

```

```python
#Problem 2 : Handling Nulls in the 'Age' Column
# Null count in 'Age'
age_nulls = first_class_df['Age'].isnull().sum()
print("Null values in 'Age':", age_nulls)

# Drop rows with null 'Age'
cleaned_df = first_class_df.dropna(subset=['Age'])
print("Rows with null 'Age' dropped.")
```

```python
#Problem 3 : One-Hot Encoding of the Embarked Column in the Titantic Dataset
# Step 1: One-hot encode the 'Embarked' column
# Changing 'titanic_df' to 'df', assuming 'df' holds the Titanic dataset
embarked_one_hot = pd.get_dummies(df['Embarked'], prefix='Embarked')

# Step 2: Add the encoded columns to the original DataFrame
titanic_encoded = pd.concat([df, embarked_one_hot], axis=1) # Changing 'titanic_df' to 'df'

# Step 3: Drop the original 'Embarked' column
titanic_encoded = titanic_encoded.drop(columns=['Embarked'])

# Step 4: Display the first few rows of the modified DataFrame
print("Modified Titanic DataFrame (with one-hot encoding for 'Embarked'):")
print(titanic_encoded.head())
```

```python
#Problem 4 : Compare Mean Survival Rates By Gender
# Assuming 'df' holds the Titanic dataset, changing 'titanic_df' to 'df'
# Calculate mean survival rates by gender
gender_survival_rates = df.groupby("Sex")["Survived"].mean() # Changing 'titanic_df' to 'df'

# Display survival rates
print("Mean Survival Rates by Gender:")
print(gender_survival_rates)

# Visualization of Survival Rates by Gender
# Assuming necessary libraries are imported
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.barplot(x=gender_survival_rates.index, y=gender_survival_rates.values, palette="muted")
plt.title("Survival Rates by Gender", fontsize=16)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Survival Rate", fontsize=12)
plt.ylim(0, 1)
plt.show()
```

```python
#Problem 5 : Visualization of Survival Rates By Port of Embarkation
import pandas as pd # Importing pandas for DataFrame operations
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' holds the Titanic dataset, changing 'titanic_df' to 'df'
# Calculate survival rates by port of embarkation
embarked_survival_rates = df.groupby("Embarked")["Survived"].mean() # Changed 'titanic_df' to 'df'

# Display survival rates
print("Survival Rates by Port of Embarkation:")
print(embarked_survival_rates)

# Visualization: Survival rates by port of embarkation
plt.figure(figsize=(8, 5))
sns.barplot(x=embarked_survival_rates.index, y=embarked_survival_rates.values, palette="coolwarm")
plt.title("Survival Rates by Port of Embarkation", fontsize=16)
plt.xlabel("Port of Embarkation", fontsize=12)
plt.ylabel("Survival Rate", fontsize=12)
plt.ylim(0, 1)
plt.show()

# Survival Rates by Gender and Port of Embarkation
grouped_survival = df.groupby(["Embarked", "Sex"])["Survived"].mean().unstack() # Changed 'titanic_df' to 'df'

# Visualization: Survival rates by port and gender
grouped_survival.plot(kind="bar", figsize=(10, 6), colormap="coolwarm", edgecolor="black")
plt.title("Survival Rates by Gender and Port of Embarkation", fontsize=16)
plt.xlabel("Port of Embarkation", fontsize=12)
plt.ylabel("Survival Rate", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Gender")
plt.show()

```
