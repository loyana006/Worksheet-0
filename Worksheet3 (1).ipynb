```python
#1.Sorting Tasks
#subseting Rows
# Task 1: Create a DataFrame with only the Fare column
fare = df[['Fare']]
print(fare.head())

# Task 2: Create a DataFrame with only the Pclass and Age columns
class_age = df[['Pclass', 'Age']]
print(class_age.head())

# Task 3: Create a DataFrame with only the Survived and Sex columns
survived_gender = df[['Survived', 'Sex']]
print(survived_gender.head())
```

```python
#2.Subsetting Rows
# Filter passengers with fare > 100
fare_gt_100 = df[df['Fare'] > 100]
print(fare_gt_100)

# Filter passengers with Pclass = 1
first_class = df[df['Pclass'] == 1]
print(first_class)

# Filter passengers who are female and under 18
female_under_18 = df[(df['Sex'] == 'female') & (df['Age'] < 18)]
print(female_under_18)
```

```python
#Subsetting Rows By Categorical Variabes
# Filter passengers who Embarked at "C" or "S"
embarked_c_or_s = df[df['Embarked'].isin(['C', 'S'])]
print(embarked_c_or_s)

# Filter passengers with Pclass in [1, 2]
first_second_class = df[df['Pclass'].isin([1, 2])]
print(first_second_class)

```

```python
#Exploratory Data Analysis Excercise 1
# Handle missing Age values by filling with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Add a new column: Fare per year
df['Fare_per_year'] = df['Fare'] / df['Age']

# Subset rows where Fare_per_year > 5
high_fare_age = df[df['Fare_per_year'] > 5]

# Sort by descending Fare_per_year
high_fare_age_srt = high_fare_age.sort_values(by='Fare_per_year', ascending=False)

# Select Name and Fare_per_year
result = high_fare_age_srt[['Name', 'Fare_per_year']]
print(result.head())
```

```python
#Highest Fare Relative To Class For Adult Males
# Add a new column: Fare per class
df['Fare_per_class'] = df['Fare'] / df['Pclass']

# Filter adult males (Age >= 18 and Sex = "male")
adult_males = df[(df['Age'] >= 18) & (df['Sex'] == 'male')]

# Sort by descending Fare_per_class
adult_males_srt = adult_males.sort_values(by='Fare_per_class', ascending=False)

# Select Name, Age, and Fare_per_class
result_males = adult_males_srt[['Name', 'Age', 'Fare_per_class']]
print(result_males.head())
```

```python
#Group-by Method Practice
# Step 1: Calculate the total fare paid across all passengers
total_fare = df['Fare'].sum()

# Step 2: Calculate the total fare for each class
fare_first_class = df[df['Pclass'] == 1]['Fare'].sum()
fare_second_class = df[df['Pclass'] == 2]['Fare'].sum()
fare_third_class = df[df['Pclass'] == 3]['Fare'].sum()

# Step 3: Combine the fare totals into a list
fare_totals = [fare_first_class, fare_second_class, fare_third_class]

# Step 4: Calculate the proportion of fare revenue by class
fare_percentages = [(fare / total_fare) * 100 for fare in fare_totals]

# Display the results
classes = ['First Class', 'Second Class', 'Third Class']
for cls, percentage in zip(classes, fare_percentages):
    print(f"{cls}: {percentage:.2f}%")

```

```python
#Percentage of Passengers by Age Group
# Step 1: Create a new column categorizing passengers into age groups
bins = [0, 18, 64, df['Age'].max()]  # Age ranges: child (<18), adult (18-64), senior (65+)
labels = ['child', 'adult', 'senior']
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Step 2: Calculate the total number of passengers
total_passengers = df['Age_group'].count()

# Step 3: Count the number of passengers in each age group
age_group_counts = df['Age_group'].value_counts()

# Step 4: Calculate the proportion of passengers in each age group
age_group_percentages = (age_group_counts / total_passengers) * 100

# Step 5: Display the proportion as a percentage
for group, percentage in age_group_percentages.items():
    print(f"{group.capitalize()}: {percentage:.2f}%")

```
